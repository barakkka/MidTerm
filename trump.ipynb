```python
import nltk
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

text = "Donald Trump will be appearing on The View tomorrow morning to discuss Celebrity Apprentice and his new book Think Like A Champion!"
tokens = nltk.word_tokenize(text)

stemmed_words = [stemmer.stem(token) for token in tokens]
print(stemmed_words)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer


lemmatizer = WordNetLemmatizer()


text = "Donald Trump will be appearing on The View tomorrow morning to discuss Celebrity Apprentice and his new book Think Like A Champion!"

tokens = nltk.word_tokenize(text)


lemmatized_words = [lemmatizer.lemmatize(token) for token in tokens]


print(lemmatized_words)
```

```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Donald Trump will be appearing on The View tomorrow morning to discuss Celebrity Apprentice and his new book Think Like A Champion!"


tokens = nltk.word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]


freq_dist = nltk.FreqDist(filtered_tokens)


wordcloud = WordCloud(background_color='grey').generate_from_frequencies(freq_dist)


plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
